we are going to write a generic program to find the roots of a equation using the Newton Rhapson method.
Here we take the function sinxcosx here we take b=-0.3 and Initial guess as 3
For Derivative purpose i took the FORWARD DIFFERENCE METHOD to calculate the derivativeand h=1e-7

```python
import math

class newton_raphson:
    def inputs(self):
        self.b=-0.3
        self.h=1e-7
        self.xroot=3

    def roots(self):
        i=0
        while i < 25:
            fn=math.sin(self.xroot)*math.cos(self.xroot) +self.b
            fstep=math.sin(self.xroot +self.h)*math.cos(self.xroot +self.h) +self.b
            fder=(fstep-fn)/self.h
            xn=self.xroot-(fn/fder)

            if abs(fn) < 1e-6:
                print("root of x is", xn)
                break
            else:
                self.xroot=xn
            i=i+1
s1=newton_raphson()
s1.inputs()
s1.roots()

```

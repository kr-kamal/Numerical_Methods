```python
Below coding program can solve the unconstrained
optimization using the gradient descent method.

• This program have three different functions. The first one for gradient descent method, second for the function evaluation and third for
computing the derivative of the function. The user who want to use your program for a new function should only need to change the second and
third functions.
• The learning rate (the constant getting multiplied with the gradient) will be allowed to change by the user.
• This program give the optimal point and optimal value of function as an output.

• This program also generate the plot of function values with iterations. Proper labeling should be done.

Testing the program for the following functions
f(x, y)=(x − 2)^2+(y − 4)^2
```

```python
import matplotlib.pyplot as plt                   #I import the matplotlib library for graphical representation

class gradient_decent :                         #I created a class having name gradient_decent

                   def gradient_descent_method(self):   #Created a method or you can say function

                                 print("Gradient Descent Method")                           #It is just to print the name of the method/so called function as mentioned in the assifnment
                                 self.a=float(input('enter the Learning rate:'))         #Just to take the input from the user as mentioned in the Assifnment Problem
                                 self.x=float(input('enter the initial x value:'))      #To take the inital Guess from the user as we are the citizen Democratic country
                                 self.y=float(input('enter the initial y value:'))     #same as above
                                 self.der_x=0.0                                        #Varibles to store the derivatite of the g(x,y) w.r.t x
                                 self.der_y=0.0                                       #Varibles to store the derivatite of the g(x,y) w.r.t y
                                 self.grad=[self.der_x,self.der_y]                   #To store the gradient in the form of list
                                 self.fxy=0.0                                       #To store the value of g(x,y)
                                 self.k=1                                          #used to count the number of iterations
                                 self.fval=[]                                     #List to store the function values at each iteration
                                 self.kval=[]                                    #List to store the iteration values

                   def function_evaluation(self):                              #created a method or you can say function as mentioned in the assifnment

                                 print("Function Evaluation")                #Just to print the name of the method/so called function as mentioned in the assifnment
                                 j=0                                        #Used for the looping purpose
                                 while j<100:                              #Welcome to the world of loops
                                  self.kval.append(self.k)                #capturing the iteration values
                                  self.fval.append(self.fxy)             #capturing the function values

        #Next condition is very heptic to understand but I tried my best to make it simple staring from very long statement i short as much it can possible

                                  if ((self.grad[0]<0.00001 and self.grad[0]>-0.00001) and (self.grad[1]<0.00001 and self.grad[1]>-0.00001)):   #Conditions of the Convergence and the Gradient to approach to Zero
                                            print("Converged in",self.k,"iterations")                   #just to print the number of iterations taken to converge
                                            print("The optimised value of x and y is:",self.x,self.y)  #just to print the optimised value of x and y
                                            print("The value of the gradient is:",self.grad)          #To print the value of the gradient at the point of convergence
                                            print("The optimised value of fn is :",self.fxy)         #To print the optimised value of the function at the point of convergence
                                            break                                                   #we don't want to run the loop further if it is converged

                                  else :                                                          #if the above condition is not satisfied then we will go for the next iteration
                                            self.x=self.x-self.a*self.grad[0]                  #formula for updation of x by using gradient descent method
                                            self.y=self.y-self.a*self.grad[1]                 #formula for updation of y by using gradient descent method
                                            print("The value of the gradient is:",self.grad)            #to print the value of the gradient at each iteration
                                            print("The value of x and y is:",self.x,self.y)            #To print the value of x and y at each iteration
                                            print("The coresponding value of fn is :",self.fxy)       #To print the value of the function at each iteration

                                  self.derivative_of_the_function()                   #Calling the method to calculate the derivative of the function so that we can use it in the updation formula and we don't require to call it again and again using the object of the class
                                  self.k=self.k+1                                    #incerase the iteration count by 1
                                  j=j+1                                             #To run the loop further

                   def derivative_of_the_function(self):                          #Created a method or you can say function as mentioned in the assifnment

                              print("Derivative of the Function")               #To print the name of the method/so called function as mentioned in the assifnment
                              self.fxy=(self.x-2)**2+(self.y-4)**2             #Mathematical expression of the function g(x,y)
                              h=0.000001                                     #Giving the value of h which is very small to calculate the derivative by using the forward difference method
                              fwd_step=((self.x+h)-2)**2+(self.y-4)**2       #Forward step calculation to calculate the derivative w.r.t x
                              self.der_x=(fwd_step-self.fxy)/h              #Formula to calculate the derivative w.r.t x with the help of forward difference method
                              fwd_step=(self.x-2)**2+((self.y+h)-4)**2     #Forward step calculation to calculate the derivative w.r.t y
                              self.der_y=(fwd_step-self.fxy)/h            #Formula to calculate the derivative w.r.t y with the help of forward difference method
                              self.grad=[self.der_x,self.der_y]          #Entering the value of the gradient in the list

                   def graphical_representation(self):                 #I created a method or you can say function as mentioned in the assifnment to plot the graph

                              print("Wait for Graphical Representation")               #Just the older thing
                              print(self.kval,self.fval)                                 #To print the iteration values and the function values so we can cross verify it
                              plt.plot(self.kval, self.fval, marker='.')                   # plotting the points
                              plt.ylabel('Function Value   ------>',color="red", fontsize=12, fontweifht='bold')    #Giving the label to y axis to make the graph more informative
                              plt.xlabel('Iteration   ------>',color="red",fontsize=12, fontweifht='bold')        #Giving the label to x axis to make the graph more informative
                              plt.ylim(-100,130)                                                                 #I give the limit so we can visualise the graph properly
                              plt.show()                                                                     #To show the graph

s1=gradient_decent()               #Creating the object of the class necessary to call the member methods or member functions of the class
s1.gradient_descent_method()        #Calling the method to take the input from the user and to initialise the variables
s1.derivative_of_the_function()    #As i called this method in the function named function_evalution but to get the initial value of the gradient i called it here omly once
s1.function_evaluation()          #Calling the method to evaluate the function and to perform the iterations
s1.graphical_representation()    #calling the method to plot the graph as mentioned in the assifnment
```

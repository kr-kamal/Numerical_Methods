we are going to write a generic program to find the roots of a equation using the bisection method.
Here we take the function (x − a)^2 +b.
Here we take a=2 b=-25 and Upper and Lower Limits 5 and -4 respectively

```python
class bisection:

      def inputs(self):

                     self.a=2
                     self.b=-25
                     self.xu=5
                     self.xl=-4

      def roots(self):

                 i=0
                 while i<25:

                    fp=(self.xu+self.a)**2+self.b
                    fn=(self.xl+self.a)**2+self.b
                    mid=(self.xu+self.xl)/2
                    fmid=(mid+self.a)**2+self.b

                    if fn*fp==0:

                              print("root ",mid)
                              break

                    if fn*fmid <=0:

                              if (fmid==0) or (fmid<0.00001 and fmid >0 ) :

                               print("root is",mid)
                               break

                              else:

                               self.xu=mid

                    if fn*fmid >=0:

                              if (fmid==0) or (fmid>-0.0001 and fmid<0):

                               print("root is",mid)
                               break

                              else:

                               self.xl=mid
                    i=i+1

s1=bisection()
s1.inputs()
s1.roots()
```

Here we take the function sinxcosx.
Here we take b=-0.3 Upper and Lower Limits as 3 and 1 respectively

```python
import math

class bisection:
    def inputs(self):
        self.b=-0.3
        self.xu=3
        self.xl=1

    def roots(self):
        i=0
        while i < 25:
            fp=math.sin(self.xu)*math.cos(self.xu) +self.b
            fn=math.sin(self.xl)*math.cos(self.xl) +self.b
            mid=(self.xu +self.xl)/2
            fmid=math.sin(mid)*math.cos(mid) +self.b


            if abs(fmid) < 1e-5:
                print("Root is", mid)
                break

            if fn*fmid < 0:
                self.xu=mid
            else:
                self.xl=mid
            i=i+1
        print("root is", mid)

s1=bisection()
s1.inputs()
s1.roots()
```
